# 公會屬地爬蟲

```python
from selenium import webdriver
import pandas as pd

#下載chromedriver，chromedriver.exe放在爬蟲檔同層
driverpath='/Users/benbilly3/Downloads/chromedriver'#74版，win的省略
browser=webdriver.Chrome(executable_path=driverpath)#模擬瀏覽器，win的打()
url='http://service.moj.gov.tw/lawer/baseSearch.aspx'
browser.get(url)
tag=browser.find_elements_by_tag_name('td')
d=[]
for i in range(len(tag)):
    d.append(tag[i].text)
browser.close()#關閉網頁瀏覽器
browser.quit()#離開網頁瀏覽器
num1=d.index('台中律師公會')
num2=(d.index('彰化律師公會'))+1
guilds=d[num1:num2]
guilds
```

# 律師姓名清冊爬蟲

新爬取各公會人員名單，再移除重複項

```python
d2=[]
for guild in guilds:
    driverpath='/Users/benbilly3/Desktop/Legaltech/chromedriver'#74版
    browser=webdriver.Chrome(executable_path=driverpath)
    url='http://service.moj.gov.tw/lawer/baseSearch.aspx'
    browser.get(url)
    guildlink=browser.find_element_by_link_text(guild)#從text找
    guildlink.click()
    tag=browser.find_elements_by_class_name('floatDiv')
    for i in range(len(tag)):
        d2.append(tag[i].text)
    browser.close()#關閉網頁瀏覽器
    browser.quit()#離開網頁瀏覽器
        
namelist=list(set(d2))#去除不同公會重複登錄名，可能一人加很多公會，以增加爬蟲效率
namelist
```

```python
#名冊儲存
li1=pd.Series(namelist)
li1.to_csv('/Users/benbilly3/Desktop/Legaltech/律師清冊.csv')
```

```python
from selenium import webdriver
import pandas as pd
import time

#讀取清冊資料
lylist=pd.read_csv('/Users/benbilly3/Desktop/Legaltech/律師清冊.csv')
lylist=lylist.T.reset_index()
lawyerset=lylist.iloc[1,:].tolist()

#批次處理，避免斷線
D=[i for i in range(0,10500,500)]
# len(lawyerset[D[0]:D[1]])
D2=D[1:]
for num1,num2 in zip(D,D2):
    lawyerset2=lawyerset[num1:num2]

# lawyerset=lawyerset[:5]#test
# lawyerset=['沈佩霖','蔡宗倫']#測試警告框

    dataset=[]
    for lawyer in lawyerset2:
        driverpath='/Users/benbilly3/Desktop/Legaltech/chromedriver'#74版
        browser=webdriver.Chrome(executable_path=driverpath)
        url='http://service.moj.gov.tw/lawer/baseSearch.aspx'
        browser.get(url)
        txtbox=browser.find_element_by_id('ctl01_txtLawyerName')#從按鈕id找
        try:
            txtbox.send_keys(lawyer )
            elelink=browser.find_element_by_id('ctl01_imgbtnSubmit')#從按鈕id找
            elelink.click()#模擬按下
            lawyername=browser.find_element_by_link_text(lawyer)#從text找
            lawyername.click()#模擬按下
            tag=browser.find_elements_by_tag_name('td')
            
            d=[]
            for i in range(len(tag)):
                d.append(tag[i].text)
                
            #將資料存到df
            sheet={}
            for a,b in zip(d[0::2],d[1::2]):
                sheet[a]=b
            sheet=pd.DataFrame(sheet,index=[(lawyerset.index(lawyer)+1)])
            dataset.append(sheet)
            browser.close()#關閉網頁瀏覽器
            browser.quit()#離開網頁瀏覽器
        except:
            browser.switch_to_alert().accept() #若跳出警告框框則接受
            browser.close()#關閉網頁瀏覽器
            browser.quit()#離開網頁瀏覽器
            
    time.sleep(5)


    finalsheet=pd.concat(dataset)
    finalsheet.to_csv('/Users/benbilly3/Desktop/Legaltech/'+str(num1)+'律師清冊.csv')
```
